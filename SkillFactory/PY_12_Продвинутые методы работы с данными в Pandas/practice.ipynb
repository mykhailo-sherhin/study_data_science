```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv')
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
melb_df['Date']
```

```python
melb_df['Date'].dt.quarter.value_counts().iloc[1]
```

```python
for col in melb_df.columns:
    if melb_df[col].nunique() < 150 and col not in ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']:
        melb_df[col] = melb_df[col].astype('category')
```

```python
melb_df.info()
```

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
melb_df.sort_values(by='AreaRatio', ascending=False, ignore_index=True).loc[1558, 'BuildingArea']
```

```python
melb_df[
    (melb_df['Type'] == 'townhouse') & 
    (melb_df['Rooms'] > 2)
].sort_values(
    by=['Rooms', 'MeanRoomsSquare'],
    ascending=[True, False],
    ignore_index=True
).loc[18,'Price']
```

```python
melb_df.groupby(by='Type')['Price'].mean()
```

```python
melb_df.groupby('Regionname')['Distance'].min().sort_values(ascending=False)
```

```python
melb_df.groupby('Regionname')['SellerG'].agg(
    		['nunique', set]
)
```

```python
melb_df.groupby('Rooms')['Price'].mean().sort_values(ascending=False)
```

```python
melb_df.groupby('Regionname')['Lattitude'].std().sort_values()
```

```python
melb_df[
    (pd.to_datetime('2017-05-01') <= melb_df['Date']) &
    (melb_df['Date'] <= pd.to_datetime('2017-09-01'))
].groupby('SellerG')['Price'].sum().sort_values()
```

```python
melb_df.groupby(['Rooms', 'Type'])['Price'].mean().unstack().round()
```

```python
melb_df.pivot_table(
    values='Price',
    index='Rooms',
    columns='Type',
    fill_value=0
).round()
```

```python
melb_df.pivot_table(
    values='Price',
    index='Regionname',
    columns='Weekend',
    aggfunc='count',
)
```

```python
melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
```

```python
melb_df.pivot_table(
    values='Price',
    index=['Method', 'Type'],
    columns='Regionname',
    aggfunc='median',
    fill_value=0
)
```

```python
pivot = melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
```
